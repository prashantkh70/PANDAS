```python
import pandas as pd
```

```python
pd.read_csv('titanic.csv')
```

```python
pd.read_csv('titanic.csv',index_col='pclass')  #index_col- to make any column as index column
```

```python
pd.read_csv('titanic.csv',header=1)  # header='row index' -To make any row index as header. Bydefualt header=0.
```

```python
pd.read_csv('titanic.csv',header=0,names=['alive','class','gender','age','sibsp','parch','price','emb','deck'])
                                                #names= defining column labels
```

```python
pd.read_csv('titanic.csv',header=0,usecols=['survived','pclass','sex','age']) 
# usecols= to select perticular columns
```

### Importing data from messy csv-files

```python
pd.read_csv('titanic_raw.csv')
```

```python
col_names=["Survived",'Class','Gender','Age','SipSp','ParCh','Fare','Enb','Deck']
```

```python
titanic=pd.read_csv('titanic_raw.csv',skiprows=3,skipfooter=2,header=None,names=col_names) 
                                                            #skiprows= to delete first three rows 
                                                            #skipfooter= to delete last two rows
                                                            #header- to label the dataframe on basis or numbers
                                                            #names- to put labels to numbers ,we need to define first.
```

```python
titanic
```

```python
titanic.to_csv("titanic_imp.csv")   # to_csv= to export as csv file
```

```python
pd.read_csv('titanic_imp.csv')
```
