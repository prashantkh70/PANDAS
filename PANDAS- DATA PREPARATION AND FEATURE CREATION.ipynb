### Discretization and Binning with pd.cut() (Part 1)

```python
#Discrete- individually separate and distinct.
#Binning- throw (something) away by putting it in a bin.
```

```python
titanic.head(10)
```

```python
age_bins=[0,10,18,30,55,100]  # alloting the bin size i.e 0 to 10, 10 to 18 -teen, 18-30,-adult,30-50 and 55 onwards
```

```python
cats=pd.cut(titanic.age,age_bins,right=False) #pd.cut- x-pass pd.Series or array with some values-column name- "age"
                                        #bins- allocated bins parameter--- specified edges of range."age_bins" 
                                        #right- Bin includes the right most edge or not i.e in 0-10 ,10 included or not
                                                #right=False exclude    True=include
```

```python
cats  # it shows the passanger in which he belongs
```

```python
cats.value_counts()# identifying how many passangers belongs to in the bin range.
```

```python
titanic["age_cat"]=cats
```

```python
titanic.head()
```

```python
titanic.groupby("age_cat").survived.mean()   # group specific survival rate  #0-10 shows highest survival rate
```

```python
group_names=["child","teenager","young_adult","adult","elderly"]   # assigned the specific names for the groups
```

```python
pd.cut(titanic.age,age_bins,right=False,labels= group_names)
```

```python
titanic["age_cat"]=pd.cut(titanic.age,age_bins,right=False, labels=group_names)  # applied all parameters in pd.cut
```

```python
titanic.head(10)
```

```python
titanic.age_cat
```

###  Scaling / Standardization- to match scale of two or more parameters.

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.fare.plot(figsize=(12,8))  # here the scales of fare column and age column are different.
titanic.age.plot(figsize=(12,8))
plt.show()
```

```python
mean_age=titanic.age.mean()    # to normalize (to match the scales) , we calculated the mean of age and fare
mean_fare=titanic.fare.mean()
```

```python
std_age=titanic.age.std()    # calculating std.deviation of age and fare.
std_fare=titanic.fare.std()
```

```python
titanic["age_z"]=(titanic.age-mean_age)/std_age   # calculated z score of fare and age column and added new columns.
titanic["fare_z"]=(titanic.fare-mean_fare)/std_fare    
```

```python
titanic.head()
```

```python
round(titanic.describe(),2)  # here the mean of age_z and fare_z are zero and std is 1. so it is get normalized.
```

```python
titanic.fare_z.plot(figsize=(12,8))  # here, the scale of both the age and fare column is same. 
                                        #standardization or scaling of features are done.
titanic.age_z.plot(figsize=(12,8))
plt.show()
```
