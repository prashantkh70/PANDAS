### Transformation with trasnform()

```python
titanic=pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.groupby(["sex","pclass"]).survived.mean()
```

```python
titanic["group_surv_rate"]=titanic.groupby(["sex","pclass"]).survived.transform("mean")
```

```python
titanic
```

```python
titanic["outliers"]=abs(titanic.survived-titanic.group_surv_rate)  #Outlier- substracted the two columns (1- 0.135447)
```

```python
titanic[titanic.outliers>0.85]  # outlier value greater than 0.85  -\
              #if the value is high ,then it is rare case that he/she has survived even though protocol was followed.  
```

### Generalizing split-apply-combine with apply()

```python
titanic=pd.read_csv("titanic.csv",usecols=["survived","pclass","sex","age","fare"])
```

```python
titanic.head()
```

```python
titanic.groupby("sex").mean()
```

```python
female_group=list(titanic.groupby("sex"))[0][1]
female_group
```

```python
female_group.mean()
```

```python
def group_mean(group):
    return group.mean()
```

```python
group_mean(female_group)
```

```python
titanic.groupby("sex").apply(group_mean)   #applying defined function on groupby method.
```

```python
titanic.nlargest(5,"age")
```

```python
def five_oldest_survived(group):                     # for each group we need oldest passanger survived 
    return group[group.survived ==1].nlargest(5,"age")   # ==1 for survived nlargest(5,"age")-
```

```python
titanic.groupby("sex").apply(five_oldest_survived)
```

### stack() and unstack()

```python
summer.head()
```

```python
medals_by_country=summer.groupby(["Country","Medal"]).Medal.count()
```

```python
medals_by_country
```

```python
medals_by_country.loc["USA"]
```

```python
medals_by_country.loc["USA","Gold"]
```

```python
medals_by_country.shape
```

```python
medals_by_country.unstack()  #we created a new column for the entries of the entries in the columns.
```

```python
medals_by_country.unstack(level=-1)  # level=-1 unstacking medal column # shows vertically- Medal has index -1
```

```python
medals_by_country.unstack(level=-2)  # level=-2 unstacking Country column # Shows Horozontally  - Country has index -2
```

```python
medals_by_country.unstack(fill_value=0)  # filling NaN values with value '0'
```

```python
medals_by_country=medals_by_country.unstack(level=-1,fill_value=0)
```

```python
medals_by_country
```

```python
medals_by_country.shape
```

```python
medals_by_country=medals_by_country[["Gold","Silver","Bronze"]]
```

```python
medals_by_country.sort_values(by=["Gold","Silver","Bronze"],ascending=[False,False,False],inplace=True)
```

```python
medals_by_country.head(10)
```

```python
medals_by_country.head(10).plot(kind="bar",figsize=(12,8),fontsize=13)
plt.xlabel("Country",fontsize=13)
plt.ylabel("Medals",fontsize=13)
plt.title("Medal per Country",fontsize=16)
plt.legend(fontsize=15)
plt.show()
```
