### Ranking DataFrame rank ()

```python
sales=pd.Series([15,32,45,21,55,15,0],index=['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
```

```python
sales
```

```python
sales.sort_values(ascending=False)
```

```python
sales.rank().sort_values(ascending=True)
```

```python
sales.rank(ascending=False, method='min').sort_values(ascending=True)
```

```python
sales.rank(ascending=False, method='average',pct=True).sort_values()   #pct is for rank in percentiles
```

### Hierarchical Indexing (Multiindex) (Part-1)

```python
titanic=pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic=titanic.iloc[:50,:]  # Selected first fifty rows and all columns
```

```python
titanic
```

```python
titanic.set_index(["pclass",'sex'],inplace=True)#selected pclass and sex as an index(multiindexing)
titanic
```

```python
titanic.sort_index()  #sorted the pclass index 
```

```python
titanic.sort_index(ascending=[False,True])  #here pclass set as desceding order and sex as ascending order.
```

```python
titanic.sort_index(ascending=[True,True])  #here pclass set as ascending order and sex as ascending order.
```

```python
titanic.swaplevel()  # swaplevel is used to swap the first index by second and vise versa.
```

```python
titanic.reset_index(inplace=True)  #reset_index -to reset the index like original one.
```

```python
titanic
```

### Hierarchical Indexing (Multiindex) (Part-2)

```python
titanic=pd.read_csv('titanic.csv')
```

```python
titanic=titanic.iloc[:50,]
```

```python
titanic
```

```python
titanic=titanic.set_index(['pclass','sex']).sort_index(ascending=True)
```

```python
titanic
```

```python
titanic.loc[1]
```

```python
titanic.loc[[1,2]]
```

```python
titanic.loc[:2]
```

```python
titanic.loc[1,'female']
```

```python
titanic.loc[(1,'female'),'age']  # selecting the multi index labels inside the bracket and passing column labels
```

```python
titanic.loc[([1,2],'female'),["age","fare"]]
```

```python
titanic.loc[([1,2],'female'),:]
```

```python
titanic.loc[(:,'female'),:]  # slicing by only using : is not allowed in pandas specially at index position.
```

```python
titanic.loc[(slice(None),slice('female')),:] # the alternative- Use slice(None) instead of : and use slice('column') 
                                                # for sub-index and then select all rows.
```

```python
titanic.loc[(slice(1),slice('female')),:]  # Here we have selected 1 for first class in pclass columns.
                                        # So in general we can use this for safety in case od multiindexing. 
```
