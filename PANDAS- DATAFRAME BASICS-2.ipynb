### Filtering Dataframes by many conditions( OR)

### Advanced filtering with between(), isin() and ~

##### between()

```python
summer=pd.read_csv('summer.csv')
summer.head()
```

```python
olygames=summer.loc[summer.Year==1988]
olygames.head()
```

```python
olygames.tail()
```

```python
olygames.info()
```

```python
olygames_since1992= summer.loc[summer.Year>=1992]
olygames_since1992.head()
```

```python
olygames_since1992.tail()
```

```python
summer.Year.between(1960,1969).head()
```

```python
olygames_60s=summer.loc[summer.Year.between(1960,1969,inclusive=True)]    #inclusive method considers all boundaries 
olygames_60s.head()                                                 # here incl.1960 and 1969 values are considered
```

##### isin()

```python
my_fav_games=[1972,1996]
my_fav_games
```

```python
summer.Year.isin(my_fav_games).head() 
# isin Method-The result will only be true at a location if all the labels match.
#here isin method will show the data only of 1972 and 1996.
```

```python
olygames_72_96=summer.loc[summer.Year.isin(my_fav_games)]
olygames_72_96.head()
```

```python
olygames_72_96.tail()
```

#####  use of ~ symbol  

```python
olygames_Not_72_96=summer.loc[~summer.Year.isin(my_fav_games)]   # '~' gives the opposite of dataframe 
                                                                # i.e values other than 1972 and 1996
```

```python
olygames_Not_72_96
```

```python
olygames_Not_72_96.Year.unique()
```

### CREATING DATAFRAMES FROM SCRATCH with pd.DataFrame()

#### Having column in place

```python
player= ['Lionel Messi','Cristiano Ronaldo','Neymar Junior','Kylian Mbappe','Manuel Neuer']
```

```python
nationality= ['Argentina','Portugal','Brazil','France','Germany']
```

```python
height= [1.70,1.87,1.75,1.78,1.93]
```

```python
goals_2018= [45,44,28,21,0]
```

#####  Having columns

```python
dic={"Player":player,"Nationality":nationality,'Club':club,"World_Champion":world_champion,"Height":height,"Goals_2018":goals_2018}
```

```python
dic
```

```python
df=pd.DataFrame(data=dic)
```

```python
df
```

```python
players=df.set_index('Player')
```

```python
players
```
